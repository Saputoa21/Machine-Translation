## Compute BLEU and chrF Scores

This is the command to compute BLEU and chrF:

sacrebleu reference-file -l language pair -i MT output -m metrics



e.g. (commands in colab start with !)

!sacrebleu en-de.vaccine.reference.de -l en-de -i en-de.opuscat.original.de -m bleu chrf

# Compute COMET Scores

Command for COMET score

comet-score -s source-file  -t MT ouput -r reference-file

e.g.

!comet-score -s *en-de.vaccine.source.en* -t *en-de.opuscat.original.de* -r *en-de.vaccine.reference.de*

-s = source,
-t = translation,
-r = reference

# QE

```python
# Unbabel/wmt22-cometkiwi-da
!pip install --upgrade pip  # ensures that pip is current
!pip install "unbabel-comet>=2.1.0"
```

```python
from comet import download_model, load_from_checkpoint

model_path = download_model("Unbabel/wmt23-cometkiwi-da-xl")
model = load_from_checkpoint(model_path)
data = [
    {
        "src": "The output signal provides constant sync so the display never glitches.",
        "mt": "Das Ausgangssignal bietet eine konstante Synchronisation, so dass die Anzeige nie strt."
    },
    {
        "src": "Krouek ilustrace je uren vem milovnkm umn ve vku od 10 do 15 let.",
        "mt": "          10  15 ."
    },
    {
        "src": "Mandela then became South Africa's first black president after his African National Congress party won the 1994 election.",
        "mt": "1994"
    }
]

model_output = model.predict(data, batch_size=8, gpus=1)
print (model_output)
```
